```python
import os
import csv

#Files to load
import_file_election = os.path.join("election_data.csv")
output_file_election = os.path.join("election_analysis.txt")


total_votes = 0

#Creating lists

candidate_options = []
candidate_votes= {}

winning_candidate = ""
winning_count= 0

#Read csv 
with open(import_file_election) as election_data:
    reader= csv.reader(election_data)
    
    #Read header row
    header = next(reader)
    #print(header)
    
    for row in reader:
        #Add total vote count
        total_votes = total_votes + 1
        
        
        #Extract candidate name and add to list
        candidate_name = row[2]
        
        if candidate_name not in candidate_options:
            
            candidate_options.append(candidate_name)
            
            #tracking vote counts of each candidate
            
            candidate_votes[candidate_name] = 0
        
        candidate_votes[candidate_name] = candidate_votes[candidate_name] + 1
        
        
with open (output_file_election, "w") as txt_file:
    
        
        
    election_results = (
        f"\nElection Results\n" +
        "======================\n" + 
        "Total Votes: " + str(total_votes)+ "\n" + 
        "======================\n"
        
    )
                
    
    print(election_results, end = "")  
             

    for candidate in candidate_votes:
        
    #Calculate vote count percentage
        votes = candidate_votes.get(candidate)
        vote_percentage = float(votes) / float (total_votes) * 100
        
        #Find winning candidate
        if (votes > winning_count):
            winning_count = votes
            winning_candidate = candidate
            
        #Print candidates voter count and percentage
        voter_output = f"{candidate}: {vote_percentage:.3f}% ({votes})\n"
        
        print(voter_outuput,end="")
        
        text_file.write(voter_ouput)
        
    #Print the winning candidates winning
    
    winning_candidate_summary = (
    
    f"======================/n"
    f" Winner: {winning_candidate}\n"
    f"======================/n"
    
    )
    
 #save winning cnadidate next to text file

txt_file.write(winning_candidate_summary)


        
            
        
    
```
